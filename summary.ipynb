```python
import pandas as pd
from pathlib import Path
import json
import re
```

```python
root = Path('test/bevformer_base/')
TEST = [*root.glob('*')]

df = {
      'regex':[],
      'mAP':[],
      'test':[]
      }

with open('include_test','r') as f:
        _include = [line.strip() for line in f.readlines()]
```

```python

for test_path in TEST:
        test_name = test_path.stem
        # print(test_name, re.match(r'^test_',test_name))
        with open(test_path/'pts_bbox'/'metrics_summary.json','r') as f:
                data = json.load(f)
                mAP = data['mean_ap']
        if re.match(r'^test_',test_name):
                regex = _include[int(re.search(r"test_(\d+)",test_name).group(1))]
                
        else:
                regex = test_name

        df['test'].append(test_name)
        df['mAP'].append(round(mAP*100,3))
        df['regex'].append(regex)
        
df = pd.DataFrame(df)
df.sort_values(by='mAP', ascending=False, inplace=True)


```
